```python
text = " ".join(text2)
text_corrected = " ".join(word for word in text2 if word.isalpha())
word_list = nltk.word_tokenize(text_corrected)
```

```python
# Створення об'єкту FreqDist для аналізу частоти слів у тексті
freq_dist = FreqDist(word_list)

# Знаходження слів-унікалінів (hapaxes), тобто слів, які зустрічаються лише один раз у тексті
hapaxes = freq_dist.hapaxes()

# Знаходження слів з низькою частотою, тобто слів, які зустрічаються менше 5 разів у тексті
low_freq_words = [word for word, freq in freq_dist.items() if freq < 5]

# Виведення результатів
print("Слова-унікаліними (hapaxes):", hapaxes)
print("Слова, які зустрічаються менше 5 разів:", low_freq_words)
```

```python
word_lengths = [len(word) for word in word_list]

length_freq_dist = FreqDist(word_lengths)

sorted_keys = sorted(length_freq_dist.keys())
sorted_values = [length_freq_dist[key] for key in sorted_keys]

plt.plot(sorted_keys, sorted_values)
plt.xticks(range(min(sorted_keys), max(sorted_keys)+1))
plt.xlabel('Довжина слова')
plt.ylabel('Частота')
plt.title('Графік довжини слів до їх частоти (відсортований)')

plt.show()
```

```python
import nltk
import re

def extract_monetary_amounts(text):
    # Знаходимо всі грошові суми у тексті за допомогою регулярного виразу
    amounts = re.findall(r'\$\d+(\.\d{2})?', text)
    return amounts

def convert_to_float(amounts):
    # Конвертуємо грошові суми в числа з плаваючою комою
    return [float(amount[1:]) if '.' in amount else int(amount[1:]) for amount in amounts if amount]

def sum_monetary_amounts(text):
    amounts = extract_monetary_amounts(text)
    amounts_float = convert_to_float(amounts)
    total_amount = sum(amounts_float)
    return total_amount

# Відкриваємо файл з вказанням правильного кодування (UTF-8)
with open('text2.txt', 'r', encoding='utf-8') as file:
    text = file.read()

# Обчислюємо загальну суму грошових сум
total_amount = sum_monetary_amounts(text)
print("Total sum of monetary amounts:", total_amount)
```
